```python
import pandas as pd
import numpy as np
import random

# Load the original CSV data
df_original = pd.read_csv('download.csv')

# Provided list of agricultural crop plants (updated and expanded)
agricultural_plants = [
    "Wheat", "Rice", "Maize", "Barley", "Oats", "Rye", "Sorghum", "Millet", "Triticale", "Fonio",
    "Teff", "Buckwheat", "Quinoa", "Amaranth", "Cassava", "Yam", "Sweet Potato", "Potato", "Taro",
    "Cocoyam", "Sugarcane", "Sugar Beet", "Oil Palm", "Soybean", "Peanut", "Sunflower", "Rapeseed",
    "Canola", "Cotton", "Flax", "Hemp", "Jute", "Kenaf", "Sisal", "Coffee", "Tea", "Cocoa", "Tobacco",
    "Banana", "Plantain", "Apple", "Pear", "Peach", "Plum", "Cherry", "Apricot", "Grapes", "Orange",
    "Lemon", "Lime", "Mandarin", "Pineapple", "Mango", "Papaya", "Guava", "Avocado", "Olive",
    "Watermelon", "Cantaloupe", "Honeydew", "Pumpkin", "Cucumber", "Zucchini", "Squash", "Tomato",
    "Eggplant", "Bell Pepper", "Chili Pepper", "Carrot", "Beetroot", "Radish", "Turnip", "Rutabaga",
    "Onion", "Garlic", "Leek", "Shallot", "Celery", "Lettuce", "Spinach", "Kale", "Cabbage", "Broccoli",
    "Cauliflower", "Brussels Sprouts", "Mustard Greens", "Collard Greens", "Okra", "Asparagus",
    "Artichoke", "Peas", "Green Beans", "Lima Beans", "Chickpeas", "Lentils", "Pigeon Pea",
    "Mung Bean", "Black Gram", "Kidney Bean", "Navy Bean", "Pinto Bean", "Black Bean", "Fava Bean",
    "Alfalfa", "Clover", "Timothy Grass", "Bermuda Grass", "Ryegrass", "Switchgrass", "Miscanthus",
    "Napier Grass", "Sesame", "Safflower", "Castor Bean", "Perilla", "Basil", "Coriander", "Parsley",
    "Dill", "Mint", "Rosemary", "Thyme", "Oregano", "Sage", "Fennel", "Chive", "Tamarind",
    "Date Palm", "Fig", "Pomegranate", "Lychee", "Rambutan", "Durian", "Mangosteen", "Longan",
    "Jackfruit", "Breadfruit", "Coconut", "Areca Nut", "Betel Leaf", "Vanilla", "Cardamom",
    "Black Pepper", "Turmeric", "Ginger", "Cinnamon", "Clove", "Nutmeg", "Allspice", "Sago Palm",
    "Arrowroot", "Horseradish", "Wasabi", "Chicory", "Endive", "Watercress", "Lotus Root", "Bamboo Shoot",
    "Agave", "Aloe Vera", "Kudzu", "Roselle", "Tinda", "Pointed Gourd", "Bitter Gourd", "Snake Gourd",
    "Ivy Gourd", "Wax Gourd", "Ash Gourd", "Bottle Gourd", "Spiny Gourd", "Yardlong Bean", "Winged Bean",
    "Indian Gooseberry", "Indian Jujube", "Tamarillo", "Chayote", "Persimmon", "Mulberry", "Elderberry",
    "Cranberry", "Blueberry", "Strawberry", "Raspberry", "Blackberry", "Gooseberry", "Currant",
    "Barberry", "Sea Buckthorn", "Loquat", "Sapodilla", "Star Apple", "Cempedak", "Salak", "Pulasan",
    "Noni", "Breadnut", "Lucuma", "Marang", "Hog Plum", "Cashew", "Macadamia", "Pecan", "Walnut",
    "Almond", "Hazelnut", "Pistachio", "Chestnut", "Brazil Nut", "Pine Nut", "Shea Nut", "Kola Nut",
    "Bamboo", "Eucalyptus", "Rubber Tree", "Neem", "Moringa", "Jatropha", "Mulberry (for sericulture)",
    "Teak", "Mahogany", "Acacia", "Albizia", "Poplar", "Willow", "Paulownia", "Casuarina",
    "Triticale", "Celeriac", "Scallion", "Sorrel", "Skirret", "Parsnip", "Lovage", "Purslane",
    "Malabar Spinach", "Amsoi", "Pak Choi", "Mizuna", "Tat Soi", "Komatsuna", "Chinese Cabbage",
    "Daikon", "Wasabina", "Arugula", "Iceberg Lettuce", "Romaine Lettuce", "Butterhead Lettuce",
    "Mâche", "Mustard Spinach", "New Zealand Spinach", "Tree Spinach", "Garden Cress", "Upland Cress",
    "Land Cress", "Water Spinach", "Swamp Cabbage", "Galangal", "Elephant Garlic", "Chinese Chive",
    "Wild Garlic", "Turk’s Turban", "Pattypan Squash", "Acorn Squash", "Delicata Squash", "Crookneck Squash",
    "Hubbard Squash", "Spaghetti Squash", "Kabocha", "Gem Squash", "Kiwano", "Indian Cucumber",
    "Thai Basil", "Holy Basil", "Vietnamese Coriander", "Shiso", "Ajwain", "Epazote", "Lemongrass",
    "Bay Leaf", "Savory", "Marjoram", "Anise", "Caraway", "Celosia", "Ube", "Duku", "Langsat",
    "Santol", "Soursop", "Sugar Apple", "Custard Apple", "Cherimoya", "Atemoya", "Jabuticaba",
    "Feijoa", "Medlar", "Mountain Apple", "Bignay", "Bilimbi", "Camu Camu", "Cocoplum", "Bacaba",
    "Mamoncillo", "Monkey Orange", "Ackee", "Jicama", "Chaya", "Nasturtium", "Ice Plant", "Sunroot",
    "Earthnut Pea", "Groundnut", "Velvet Bean", "Hyacinth Bean", "Wingpod Purslane", "Cluster Fig",
    "Yellowhorn", "Indian Laurel", "Kaffir Lime", "Burmese Grape", "Indian Almond", "Sea Grape",
    "Chinese Toon", "Bitterleaf", "Tindora", "Corn Salad", "Indian Sorrel", "Mountain Yam"
]


# Get existing labels from the original DataFrame and convert to lowercase set for efficient lookup
existing_labels_lower = {label.lower() for label in df_original['label'].unique()}

# Filter agricultural_plants to get names not already in original data
available_new_plant_names = [
    plant for plant in agricultural_plants if plant.lower() not in existing_labels_lower
]

# Number of entries to create for synthetic data
num_needed_new_entries = 500 - len(df_original)
if num_needed_new_entries < 0:
    num_needed_new_entries = 0 # No new entries needed if original is already 500+

# Select new labels randomly from the available names. If more needed than available unique, some will be repeated.
if num_needed_new_entries > len(available_new_plant_names):
    print(f"Warning: Not enough unique plant names ({len(available_new_plant_names)}) in the provided list to reach {num_needed_new_entries} new entries. Repeating names.")
    selected_new_labels = random.choices(available_new_plant_names, k=num_needed_new_entries)
else:
    selected_new_labels = random.sample(available_new_plant_names, k=num_needed_new_entries)

# Calculate statistics for numerical columns from original data
ranges = {}
for col in ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']:
    if col in df_original.columns:
        ranges[col] = {
            'mean': df_original[col].mean(),
            'std': df_original[col].std(),
            'min': df_original[col].min(),
            'max': df_original[col].max()
        }
    else:
        # Fallback values if a column is unexpectedly missing, though highly unlikely here.
        print(f"Warning: Column '{col}' not found in original DataFrame. Using default fallback values.")
        ranges[col] = {'mean': 50, 'std': 10, 'min': 0, 'max': 100}


# Create a list to store synthetic data
synthetic_data = []

# Generate synthetic data for the new plants using the calculated statistics
for plant_name in selected_new_labels:
    new_row = {'label': plant_name}
    for col, r in ranges.items():
        # Generate values using a normal distribution
        value = np.random.normal(r['mean'], r['std'])
        # Clip values to the observed min/max to ensure realism
        new_row[col] = np.clip(value, r['min'], r['max'])
    synthetic_data.append(new_row)

# Create a DataFrame from the synthetic data
df_synthetic = pd.DataFrame(synthetic_data)

# Ensure the synthetic DataFrame has the same columns as the original (including order)
df_synthetic = df_synthetic.reindex(columns=df_original.columns)

# Combine the original plants and synthetic data
df_combined = pd.concat([df_original, df_synthetic], ignore_index=True)

# Shuffle the combined DataFrame and reset the index for a clean new index
df_combined = df_combined.sample(frac=1, random_state=42).reset_index(drop=True) # Added random_state for reproducibility

# Display the head of the combined DataFrame
print("\nCombined DataFrame Head:")
print(df_combined.head())

# Display the tail of the combined DataFrame to see synthetic data
print("\nCombined DataFrame Tail:")
print(df_combined.tail())

# Display information about the combined DataFrame
print("\nCombined DataFrame Info:")
print(df_combined.info())

# Display the shape of the combined DataFrame
print(f"\nCombined DataFrame Shape: {df_combined.shape}")
```
